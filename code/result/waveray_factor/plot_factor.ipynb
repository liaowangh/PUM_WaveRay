```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
def plot_save(wave_number, factors, labels, output_folder, output_name):    
    fig,ax = plt.subplots()
    for i in range(len(factors)):
        ax.plot(wave_number, factors[i], label = labels[i])
#     ax.plot(wave_number, factors, '^-', label="plan wave") # triangle marker
#     ax.loglog(ndofs, H1err, 's-', label=h1_label) # square marker
    ax.set_xlabel('k(wave number)')
    ax.set_ylabel('convergence factor')
#     ax.set_title(title)
    ax.legend()
    fig.savefig(output_folder + output_name +'.eps', format='eps')
```

```python
folder = "./"
file = ["epum_planwave_5541.txt", "epum_planwave_5531.txt", "epum_planwave_5542.txt", "epum_planwave_6641.txt"]
for f in file:
    sol_name = f.split('.')[0]
    df = pd.read_csv(folder + f, sep=' ')
    wave_number = df['k'].to_numpy()
    
    square = df['square'].to_numpy()
    square_hole = df['square_hole'].to_numpy()
    triangle_hole = df['triangle_hole'].to_numpy()
    
    factors = [square, square_hole, triangle_hole]
    labels = ['square', 'square_hole', 'triangle_hole']
    plot_save(wave_number, factors, labels, folder, sol_name+'no_title')
```
